```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler

class DataProcessor:
    def __init__(self):
        self.df = None
        self.train = None
        self.test = None
        self.scaler = None

    def read_data(self):
        self.df = pd.read_csv('data/cleaned.csv')
        self.df.drop(['Unnamed: 0'], axis=1, inplace=True)
        split_index = self.df.columns.get_loc("2400")
        self.df = self.df.iloc[:, :split_index]

    def select_batches(self):
        batches = np.random.choice(self.df['Batch ID'].unique(), 9, replace=False)
        self.train = self.df.loc[~self.df['Batch ID'].isin(batches)]
        self.test = self.df.loc[self.df['Batch ID'].isin(batches)]

    def preprocess_data(self):
        self.df = self.df.rename(columns={'2-PAT control(PAT_ref:PAT ref)': 'Batch reference(Batch_ref:Batch ref)','Batch reference(Batch_ref:Batch ref)':'2-PAT control(PAT_ref:PAT ref)'})
        self.df.drop(columns=[ "Batch ID","Fault flag", 'Fault flag',
        'Fault reference(Fault_ref:Fault ref)', ' 1-Raman spec recorded',
        '0 - Recipe driven 1 - Operator controlled(Control_ref:Control ref)',
        '1- No Raman spec',
        '2-PAT control(PAT_ref:PAT ref)', 'Batch reference(Batch_ref:Batch ref)',
        ], inplace=True)

    def scale_df(self):
        scaler = StandardScaler()
        scaler.fit(self.train.values)
        self.train = self._scale_df(self.train, scaler)
        self.test = self._scale_df(self.test, scaler)

    def _scale_df(self, df, scaler):
        data = scaler.transform(df.values)
        return pd.DataFrame(data, index=df.index, columns=df.columns)

    def inverse_transform(self, data):
        return self.scaler.inverse_transform(data)


data_processor = DataProcessor()
data_processor.read_data()
data_processor.select_batches()
data_processor.preprocess_data()
data_processor.scale_df()
```
